```python
import pandas as pd
import numpy as np
import seaborn as sns
import os 
pwd = os.getcwd()
```

```python
detail_path = os.path.join(pwd,'detail.csv')
price_path = os.path.join(pwd,'price.csv')
name_path  = os.path.join(pwd,'name.csv')
```

```python
detail_df = pd.read_csv(detail_path)
```

```python
price_df_raw = pd.read_csv(price_path)
price_df = pd.DataFrame(price_df_raw.columns,columns=['Price'])
```

```python
name_df_raw = pd.read_csv(name_path,sep='/')
name_df = pd.DataFrame(name_df_raw.columns,columns=["name"])
```

```python
price_name_df = name_df.join(price_df,how='outer')
price_name_df['name'] = price_name_df['name'].str.rstrip(',')
```

```python
df_merge = pd.merge(left = price_name_df, right = detail_df, how ="outer", left_on = ["name"], right_on = ["name"])
df_merge.dropna(inplace=True)
df_merge['Price'] = df_merge['Price'].str.rstrip('( ₫.1').str.rstrip(' ₫.2').str.rstrip('( ₫').str.replace('.', '', regex=True).astype(int)
df_merge['genre'] = df_merge['genre'].str.lstrip('trái cây,')
df_merge['number_review'] = df_merge['number_review'].str.lstrip('(Xem ').str.rstrip(' đánh giá)').astype(int)
df_merge['sold_number'] = df_merge['sold_number'].str.lstrip('Đã bán ')
df_merge.drop(index=8,inplace=True)
```

THE AVERAGE PRICE 

```python
df_merge['Price'].mean()
```

DOES BOOKS WITH HIGH PRICE TEND TO BE MORE POPULAR ?

```python
sns.regplot(df_merge['number_review'],df_merge['Price']/1000)
```

WHICH TYPE OF BOOK HAS THE HIGHEST PRICE

```python
genre_price = df_merge.groupby("genre")["Price"].mean().reset_index().sort_values(by="Price", ascending=False)
genre_price.plot.bar(x="genre")
```

WHICH TYPE OF BOOK HAS THE MOST REVIEWS

```python
genre_review = df_merge.groupby("genre")["number_review"].mean().reset_index().sort_values(by="number_review", ascending=False)
genre_review.plot.bar(x="genre")
```
